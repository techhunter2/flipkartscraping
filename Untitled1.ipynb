```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
link= "https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"
page= requests.get(link)
data= bs(page.content,'html.parser')
```

```python
namel=[]
pricel=[]
rating=[]
images=[]
desc=[]
for n in (data.find_all('div',class_ ="_4rR01T")):
  namel.append(n.text)
for p in (data.find_all('div',class_="_30jeq3 _1_WHN1")):
  pricel.append(p.text)
for r in data.find_all('div',class_="_3LWZlK"):
  rating.append(r.text)
for img in data.find_all('img',class_="_396cs4 _3exPp9"):
  images.append(img.get('src'))
for d in data.find_all('div', class_="fMghEO"):
  desc.append(d.text)
```

```python
rating=rating[0:24]
df = {'Name':namel,'Price':pricel,'Price':pricel,'Rating':rating,'Descriptio':desc,'Image':images}
```

```python
new = pd.DataFrame.from_dict(df)
```
